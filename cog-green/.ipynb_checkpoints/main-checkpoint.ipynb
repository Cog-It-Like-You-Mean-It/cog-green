```python
import numpy as np
import os
import io
import matplotlib.pyplot as plt
```

```python
images = []
labels = []
```

```python
def load_images(directory, label):
    for filename in os.listdir(directory):
        file_path = directory + "/" + filename
        image = plt.imread(file_path)
        images.append(image)
        labels.append(label)
```

```python
load_images("./data/trash", 0)
load_images("./data/paper-recycling", 1)
load_images("./data/other-recycling", 2)
```

```python
np.save("images.npy", np.array(images))
np.save("labels.npy", np.array(labels))
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
idxs = np.arange(len(labels))
np.random.shuffle(idxs)
train_indices = idxs[0: int(0.75 * len(labels))]
test_indices = idxs[int(0.75 * len(labels)):]
x_train = images[train_indices]
y_train = labels[train_indices]
x_test = images[test_indices]
y_test = labels[test_indices]
```

```python
def normalize(data):
    avg = data.mean(axis=0)
    std = data.std(axis=0)
    return (data - avg) / std
```

```python
normed_x_train = normalize(x_train)
normed_x_test = normalize(x_test)
```
